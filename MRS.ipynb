```python
#Now i want to get all the genres without extra data
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    c = 0
    for i in ast.literal_eval(obj):
        if c != 3:
            
            L.append(i['name'])
            c = c+1
        else:
            break
            
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_dir(obj):
    L = []
    #c = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            
            L.append(i['name'])
            break;
            
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast_x'] = movies['cast_x'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew_x'] = movies['crew_x'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#look up sklearn documentation to learn and write about this specific library

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words = 'english')
```

```python
#here we are creating the vectors of 5000 words and their presence in each movie tags
vectors = cv.fit_transform(new_df['tags']).toarray()
#also search for toarray() function and learn
#and converted into an array
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    #the idea is to first search for the movie id(index in data set)
    #based on movie
    #then search for top 5 similar movies
    movie_index = new_df[new_df['title'] == movie].index[0]
    #now i need to get top 5 similar movies
    #we can generally sort the similarity array
    #but the problem is we loose the index of the movies that way
    #so we are tring to preserve the index of the movie in the list
    #we do this by converting the array into enumerate and reconverting 
    #then into list
    distances = similarity[movie_index]
    #this return the indes of the movies along with the similarity score
    
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        #prints movie indexes
        print(i[0])
        #prints the movie name
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
